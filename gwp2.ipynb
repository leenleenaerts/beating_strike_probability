```python
import pandas_datareader.data as web
import pandas_datareader as pdr
import pandas as pd
import seaborn as sns
import yfinance as yfin
import datetime
import numpy as np
import matplotlib.pyplot as plt


yfin.pdr_override()

# Using code from FRED API: Get US Economic Data using Python
def get_fred_data(param_list, start_date, end_date):
    df = pdr.DataReader(param_list, "fred", start_date, end_date)
    return df.reset_index()
```

```python
# 1. Large Bank Consumer Credit Card Balances: Current Credit Limit: 90th Percentile
series = "RCCCBLIMPCT90"
# data for series
ninety_per_cl = get_fred_data(param_list=[series], start_date="2012-07-01", end_date="2023-06-27")
ninety_per_cl = ninety_per_cl.rename(columns={"DATE": "Date", "RCCCBLIMPCT90": "Credit Limit in USD"})


# plot a time series plot

sns.lineplot(x="Date", y="Credit Limit in USD", data=ninety_per_cl)
plt.title('Large Bank Consumer Credit Card Balances: Current Credit Limit: 90th Percentile', y=1, fontsize = 10)
# The below graph depicts the 90th percentile credit limits in the USA between the 1st of July 2012 and the 27th of June 2023.
```

```python
# 5. Spread between 2 Yr Treasury and 2-Year High Quality Market (HQM) Corporate Bond Par Yield

series1 = "HQMCB2YRP" # 2-Year High Quality Market (HQM) Corporate Bond Par Yield
series2 = "DGS2" # Market Yield on U.S. Treasury Securities at 2-Year Constant Maturity, Quoted on an Investment Basis
# data for series
corp_bond = get_fred_data(param_list=[series1], start_date="2018-07-07", end_date="2023-06-27")
corp_bond = corp_bond.rename(columns={"DATE": "Date", "HQMCB2YRP": "Yield"})
treas = get_fred_data(param_list=[series2], start_date="2018-07-07", end_date="2023-06-27")
treas = treas.rename(columns={"DATE": "Date", "DGS2": "Yield"})

# plot a time series plot
sns.lineplot(x="Date", y="Yield", data=corp_bond, label="2 Year Corporate Bond")
sns.lineplot(x="Date", y="Yield", data=treas, label="2 Year US Treasury")
plt.title('Yields', y=1, fontsize = 10)


# The below graph depicts the spread between 2 Yr Treasury and 2-Year High Quality Market (HQM) Corporate Bond Par Yield
```
